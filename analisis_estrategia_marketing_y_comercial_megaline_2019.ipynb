1. Pago mensual: $20.

3. Si se exceden los límites del paquete:
  * 1 minuto: 3 centavos.
  * 1 SMS: 3 centavos.
  * 1 GB de datos: $10.

### Ultimate

1. Pago mensual: $70.

2. 3000 minutos al mes, 1000 SMS y 30 GB de datos.

3. Si se exceden los límites del paquete:
  * 1 minuto: 1 centavo.
  * 1 SMS: 1 centavo.
  * 1 GB de datos: $7.

15. La tabla messages (datos sobre los SMS):
16. id: identificador único del SMS.
17. message_date: fecha del SMS.
18. user_id: el identificador del usuario que manda el SMS.
19. La tabla internet (datos sobre las sesiones web):
20. id: identificador único de la sesión.
21. mb_used: el volumen de datos gastados durante la sesión (en megabytes).
22. session_date: fecha de la sesión web.
23. user_id: identificador del usuario.
24. La tabla plans (datos sobre las tarifas):
25. plan_name: nombre de la tarifa.
26. usd_monthly_fee: pago mensual en dólares estadounidenses.
27. minutes_included: minutos incluidos al mes.
28. messages_included: SMS incluidos al mes.
29. mb_per_month_included: datos incluidos al mes (en megabytes).
30. usd_per_minute: precio por minuto tras exceder los límites del paquete (por ejemplo, si el paquete incluye 100 minutos, el operador cobrará el minuto 101).
31. usd_per_message: precio por SMS tras exceder los límites del paquete.
32. usd_per_gb: precio por gigabyte de los datos extra tras exceder los límites del paquete (1 GB = 1024 megabytes).

[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

```python
# Calcula el ingreso mensual para cada usuario

# Se crean nuevas columnas donde se realizaran los calculos rubro por rubro, sobre los servicios ofertados en el manual de operaciones,
# se realiza el calculo de los servicios utilizados en el mes, contra los servicios incluidos en el plan tarifario correspondiente:

master_df_including_everything['excess_minutes'] = (master_df_including_everything['total_minutes'] - master_df_including_everything['minutes_included']).clip(lower=0)
master_df_including_everything['excess_messages'] = (master_df_including_everything['total_messages'] - master_df_including_everything['messages_included']).clip(lower=0)
master_df_including_everything['excess_mb'] = (master_df_including_everything['total_internet'] - master_df_including_everything['mb_per_month_included']).clip(lower=0)

# Se crean nuevas columnas donde se realizaran los calculos para obtener los cargos excedentes de manera mensual por mes, por usuario
# con base a los consumos generados despues de los serviios incluidos en el plan tarifario:

master_df_including_everything['charge_minutes'] = master_df_including_everything['excess_minutes'] * master_df_including_everything['usd_per_minute']
master_df_including_everything['charge_messages'] = master_df_including_everything['excess_messages'] * master_df_including_everything['usd_per_message']
master_df_including_everything['charge_mb'] = np.ceil(master_df_including_everything['excess_mb'] / 1024) * master_df_including_everything['usd_per_gb']

# Se crea columna con los ingresos mensuales por mes, por usuario:

master_df_including_everything['total_monthly_revenue'] = master_df_including_everything['charge_minutes'] + master_df_including_everything['charge_messages'] + master_df_including_everything['charge_mb'] + master_df_including_everything['usd_monthly_pay']


# Para fines practicos, se creara una función donde englobe todo el proceso efectuado en este momento, para así, en lo sucesivo
# y en proximos meses, solo se tenga que correr la función para lograr obtener los resultados deseados:

def calculate_monthly_revenue(df):
        # Se crean nuevas columnas donde se realizaran los calculos rubro por rubro, sobre los servicios ofertados en el manual de operaciones,
        # se realiza el calculo de los servicios utilizados en el mes, contra los servicios incluidos en el plan tarifario correspondiente:

        df['excess_minutes'] = (df['total_minutes'] - df['minutes_included']).clip(lower=0)
        df['excess_messages'] = (df['total_messages'] - df['messages_included']).clip(lower=0)
        df['excess_mb'] = (df['total_internet'] - df['mb_per_month_included']).clip(lower=0)

        # Se crean nuevas columnas donde se realizaran los calculos para obtener los cargos excedentes de manera mensual por mes, por usuario
        # con base a los consumos generados despues de los serviios incluidos en el plan tarifario:

        df['charge_minutes'] = df['excess_minutes'] * df['usd_per_minute']
        df['charge_messages'] = df['excess_messages'] * df['usd_per_message']
        df['charge_mb'] = np.ceil(df['excess_mb'] / 1024) * df['usd_per_gb']

        # Se crea columna con los ingresos mensuales por mes, por usuario:

        df['total_monthly_revenue'] = df['charge_minutes'] + df['charge_messages'] + df['charge_mb'] + df['usd_monthly_pay']

        return df

# Verificando que el df se encuentre en orden y que los calculos se hayan efectuado tomando en cuenta las columnas en cada fila:

master_df_including_everything.info()
print(" ")
print("________________________________________________________________")
print(" ")
master_df_including_everything.head(20)

```
